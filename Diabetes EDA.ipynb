### Answering Questions from The Data

1. Are patients with high blood pressure and high cholesterol more likely to also have diabetes?

```python
df_bp_chol = df[(df.HighBP == 1) & (df.HighChol == 1)]
diabetes_count = df_bp_chol[(df_bp_chol['Diabetes_012'] == 1) | (df_bp_chol['Diabetes_012'] == 2)].shape[0]
no_diabetes_count = df_bp_chol[(df_bp_chol['Diabetes_012'] == 0)].shape[0]
total_count = df_bp_chol.shape[0]

diabetes_percent = (diabetes_count / total_count) * 100
no_diabetes_percent = (no_diabetes_count / total_count) * 100

print(round(diabetes_percent,2), '% of patients with high blood pressure and high cholestorol also have diabetes.')
print(round(no_diabetes_percent,2), 
      '% of patients with high blood pressure and high cholestorol do not also have diabetes.')
```

2. Does Diabetes affect more women or men?

```python
df_diabetes = df[(df['Diabetes_012'] == 1) | (df['Diabetes_012'] == 2)]
female_count = df_diabetes[(df_diabetes['Sex'] == 0)].shape[0]
male_count = df_diabetes[(df_diabetes['Sex'] == 1)].shape[0]
total_count = df_diabetes.shape[0]

female_percent = (female_count / total_count) * 100
male_percent = (male_count / total_count) * 100

print(round(female_percent,2), '% of patients with diabetes are female.')
print(round(male_percent,2), '% of patients with diabetes are male.')
```

3. Does diabetes affect more patients with poor health? 

```python
df_poor_health = df[(df['PhysActivity'] == 0) & # no physical activity in the last 30 days
                    (df['HvyAlcoholConsump'] == 1) & # heavy alcohol consumption
                    ((df['GenHlth'] == 4) | (df['GenHlth'] == 5)) & # general health of fair or poor
                    (df['PhysHlth'] > 10) & # physical health was not good for more than 10 days in the month 
                    (df['DiffWalk'] == 1)] # serious difficulty walking or climbing stairs

diabetes_count = df_poor_health[(df_poor_health['Diabetes_012'] == 1) | (df_poor_health['Diabetes_012'] == 2)].shape[0]
total_count = df_poor_health.shape[0]

diabetes_percent = (diabetes_count / total_count) * 100

print(round(diabetes_percent,2), '% of patients with poor health also have diabetes.')
```

```python
df_good_health = df[(df['PhysActivity'] == 1) & # physical activity in the last 30 days
                    (df['HvyAlcoholConsump'] == 0) & # no heavy alcohol consumption
                    ((df['GenHlth'] == 1) | (df['GenHlth'] == 2) | (df['GenHlth'] == 3)) & # excellent/very good/good
                    (df['PhysHlth'] < 10) & # physical health was not good for less than 10 days in the month 
                    (df['DiffWalk'] == 0)] # no serious difficulty walking or climbing stairs

diabetes_count = df_good_health[(df_good_health['Diabetes_012'] == 1) | (df_good_health['Diabetes_012'] == 2)].shape[0]
total_count = df_good_health.shape[0]

diabetes_percent = (diabetes_count / total_count) * 100

print(round(diabetes_percent,2), '% of patients with good health have diabetes.')
```

4. Do paitents with lower incomes have a higher chance of diabetes?

```python
df_low_income = df[(df['AnyHealthcare'] == 0) & # no healthcare coverage
                    (df['NoDocbcCost'] == 1) & # needed to see a doc but couldnt because of cost
                    (df['Income'] < 8) & # income less than $75K
                    (df['Education'] < 5)] # no college education

diabetes_count = df_low_income[(df_low_income['Diabetes_012'] == 1) | (df_low_income['Diabetes_012'] == 2)].shape[0]
total_count = df_low_income.shape[0]

diabetes_percent = (diabetes_count / total_count) * 100

print(round(diabetes_percent,2), '% of low income patients have diabetes.')
```

```python
df_high_income = df[(df['AnyHealthcare'] == 1) & # healthcare coverage
                    (df['NoDocbcCost'] == 0) & # never needed to see a doc but couldnt because of cost
                    (df['Income'] >= 8) & # income greater than $75K
                    (df['Education'] > 4)] # college educated

diabetes_count = df_high_income[(df_high_income['Diabetes_012'] == 1) | (df_high_income['Diabetes_012'] == 2)].shape[0]
total_count = df_high_income.shape[0]

diabetes_percent = (diabetes_count / total_count) * 100

print(round(diabetes_percent,2), '% of high income patients have diabetes.')
```
